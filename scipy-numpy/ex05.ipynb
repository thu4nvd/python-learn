```python
import pandas as pd
import matplotlib as plt
```

```python
df = pd.read_csv(
    'dientich.csv',
    index_col=0,
    names=['Tên','2011','2012','2013','2014','2015'],
    delimiter=';'
)
df
```

```python
df.columns[4]
```

```python
df.sort_values(by=['2015'], ascending=False, inplace=True, na_position='last')
df
```

```python
df_max10 = df[6:16]
df_max10
```

```python
plt.rc('font', family='Arial')
```

```python
df_max10.plot.bar()
```
